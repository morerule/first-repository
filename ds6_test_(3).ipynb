```python
from sklearn.preprocessing import StandardScaler

df_copy = train_df.copy()
df_copy['Hour'] = (df_copy['Time'] // 3600) % 24
df_copy['Hour_sin'] = np.sin(2 * np.pi * df_copy['Hour'] / 24)
df_copy['Hour_cos'] = np.cos(2 * np.pi * df_copy['Hour'] / 24)
df_copy['TimeDiff'] = df_copy['Time'].diff().fillna(0)

# Time 피처를 위한 Scaler를 생성하고 학습시킵니다.
time_scaler = StandardScaler()
time_scaler.fit(df_copy[['Time']])
df_copy['Time_scaled'] = time_scaler.transform(df_copy[['Time']])
df_copy = df_copy.drop(['id', 'Time'], axis=1)

new_cols = ['Hour', 'Hour_sin', 'Hour_cos', 'TimeDiff', 'Time_scaled']
other_cols = [col for col in df_copy.columns if col not in new_cols]
df_copy = df_copy[new_cols + other_cols]

print(df_copy.head())
```

```python
df_copy.describe( )
```

```python
def get_train_test_dataset(df):
    X_features = df.iloc[:, :-1]
    y_target = df.iloc[:, -1]
    X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.3, random_state=0, stratify=y_target)
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = get_train_test_dataset(df_copy)
```

```python
print('학습 데이터 레이블 값 비율')
print(y_train.value_counts()/y_train.shape[0] * 100)
print('테스트 데이터 레이블 값 비율')
print(y_test.value_counts()/y_test.shape[0] * 100)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score

def get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix( y_test, pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test, pred_proba)
    print('오차 행렬')
    print(confusion)
    print('정밀도: {0:.4f}, 재현율: {1:.4f},\
    F1: {2:.4f}, AUC:{3:.4f}'.format(precision, recall, f1, roc_auc))
```

```python
from sklearn.linear_model import LogisticRegression

lr_clf = LogisticRegression()
lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
lr_pred_proba = lr_clf.predict_proba(X_test)[:, 1]

get_clf_eval(y_test, lr_pred, lr_pred_proba)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# 예측 확률 기반으로 ROC Curve 좌표 계산
fpr, tpr, thresholds = roc_curve(y_test.values, lr_pred_proba)

# AUC 점수 계산
roc_auc = roc_auc_score(y_test, lr_pred_proba)

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.4f})', color='blue')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # 기준선
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.title('ROC Curve - Logistic Regression')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
def get_model_train_eval(model, ftr_train=None, ftr_test=None, tgt_train=None, tgt_test=None):
    model.fit(ftr_train, tgt_train)
    pred = model.predict(ftr_test)
    pred_proba = model.predict_proba(ftr_test)[:, 1]
    get_clf_eval(tgt_test, pred, pred_proba)
```

```python
import numpy as np

def get_outlier(df=None, column=None, weight=1.5):
    # target이 1인 데이터만 필터링
    fraud = df[df['Class']==1][column]
    quantile_25 = np.percentile(fraud.values, 25)
    quantile_75 = np.percentile(fraud.values, 75)
    # IQR 계산
    iqr = quantile_75 - quantile_25
    iqr_weight = iqr * weight
    lowest_val = quantile_25 - iqr_weight
    highest_val = quantile_75 + iqr_weight
    # 이상치 인덱스 추출
    outlier_index = fraud[(fraud < lowest_val) | (fraud > highest_val)].index
    return outlier_index
```

```python
outlier_index_v17 = get_outlier(df=df_log, column='V17', weight=1.5)
print('V17 이상치 데이터 인덱스:', outlier_index_v17)
```

```python
def get_preprocessed_df_outlier(df):
    df_temp = df.copy()
    outlier_index = get_outlier(df=df_temp, column='V14', weight=1.5)
    df_temp.drop(outlier_index, axis=0, inplace=True)
    return df_temp

df_outlier = get_preprocessed_df_outlier(df_log)
X_train, X_test, y_train, y_test = get_train_test_dataset(df_outlier)

print('### 로지스틱 회귀 예측 성능 (이상치 제거 후) ###')
lr_clf = LogisticRegression(max_iter=1000)
get_model_train_eval(lr_clf, ftr_train=X_train, ftr_test=X_test, tgt_train=y_train, tgt_test=y_test)

print('### LightGBM 예측 성능 (이상치 제거 후) ###')
lgbm_clf = LGBMClassifier(n_estimators=1000, num_leaves=64, n_jobs=-1, boost_from_average=False)
get_model_train_eval(lgbm_clf, ftr_train=X_train, ftr_test=X_test, tgt_train=y_train, tgt_test=y_test)

print('### XGBoost 예측 성능 (이상치 제거 후) ###')
xgb_clf = XGBClassifier(n_estimators=1000, learning_rate=0.1, max_depth=3, random_state=0, eval_metric='logloss')
get_model_train_eval(xgb_clf, ftr_train=X_train, ftr_test=X_test, tgt_train=y_train, tgt_test=y_test)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0)
X_train_over, y_train_over = smote.fit_resample(X_train, y_train)
print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트: ', X_train.shape, y_train.shape)
print('SMOTE 적용 후 학습용 피처/레이블 데이터 세트: ', X_train_over.shape, y_train_over.shape)
print('SMOTE 적용 후 레이블 값 분포: \n', pd.Series(y_train_over).value_counts())
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.metrics import precision_recall_curve

def precision_recall_f1_plot(y_test, pred_proba_c1):
    precisions, recalls, thresholds = precision_recall_curve(y_test, pred_proba_c1)
    f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-6)

    # 최적 F1 점수와 해당 threshold 찾기
    best_index = np.argmax(f1_scores[:-1])
    best_threshold = thresholds[best_index]
    best_f1 = f1_scores[best_index]

    # 그래프 그리기
    plt.figure(figsize=(8,6))
    plt.plot(thresholds, precisions[:-1], linestyle='--', label='Precision', color='blue')
    plt.plot(thresholds, recalls[:-1], label='Recall', color='red')
    plt.plot(thresholds, f1_scores[:-1], label='F1 Score', color='green')
    plt.axvline(x=best_threshold, color='gray', linestyle=':', label=f'Best F1 Threshold = {best_threshold:.2f}')

    plt.xlabel('Threshold')
    plt.ylabel('Score')
    plt.title('Precision, Recall, and F1 Score vs Threshold')
    plt.legend()
    plt.grid()
    plt.xticks(np.round(np.arange(thresholds.min(), thresholds.max(), 0.1), 2))
    plt.show()

    # 수치 출력
    print(f'최고 F1 점수: {best_f1:.4f} at Threshold = {best_threshold:.2f}')
```

```python
from sklearn.model_selection import GridSearchCV

print('\n### 로지스틱 회귀 하이퍼파라미터 튜닝 (L1, L2 규제) ###')

# 튜닝할 하이퍼파라미터 그리드 설정
params = {
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.1, 1, 10, 100] # C는 규제 강도의 역수. 작을수록 강한 규제.
}

# Logistic Regression 모델 생성 (L1 규제를 위해 solver 변경)
lr_clf_tune = LogisticRegression(solver='liblinear', max_iter=1000)

# GridSearchCV를 사용하여 최적의 하이퍼파라미터 탐색
grid_cv_lr = GridSearchCV(lr_clf_tune, param_grid=params, cv=3, scoring='f1', verbose=1, n_jobs=-1)
grid_cv_lr.fit(X_train_over, y_train_over)

print('GridSearchCV 최적 하이퍼파라미터 (Logistic Regression):', grid_cv_lr.best_params_)
print('GridSearchCV 최고 F1 점수 (Logistic Regression):', grid_cv_lr.best_score_)

# 최적 하이퍼파라미터로 모델 재학습 및 평가
best_lr = grid_cv_lr.best_estimator_

print('\n### 최적 하이퍼파라미터 로지스틱 회귀 예측 성능 (SMOTE) ###')
get_model_train_eval(best_lr, ftr_train=X_train_over, ftr_test=X_test, tgt_train=y_train_over, tgt_test=y_test)

print('\n### 최적 하이퍼파라미터 로지스틱 회귀 F1 스코어 (SMOTE) ###')
best_lr_pred_proba = best_lr.predict_proba(X_test)[:, 1]
precision_recall_f1_plot(y_test, best_lr_pred_proba)
```

```python
print('\n### LightGBM 하이퍼파라미터 튜닝 ###')
# 하이퍼파라미터 튜닝을 위한 LGBM 모델 생성
lgbm_clf_tune = LGBMClassifier(boost_from_average=False)

# 튜닝할 하이퍼파라미터 그리드 설정 (시간을 고려하여 일부만 테스트)
params = {
    'n_estimators': [200, 500],
    'num_leaves': [32, 64],
    'max_depth': [10, 20],
    'learning_rate': [0.05, 0.1]
}

# GridSearchCV를 사용하여 최적의 하이퍼파라미터 탐색
# SMOTE로 오버샘플링된 데이터를 사용합니다.
grid_cv_lgbm = GridSearchCV(lgbm_clf_tune, param_grid=params, cv=3, scoring='f1', verbose=1, n_jobs=-1)
grid_cv_lgbm.fit(X_train_over, y_train_over)

print('GridSearchCV 최적 하이퍼파라미터 (LGBM):', grid_cv_lgbm.best_params_)
print('GridSearchCV 최고 F1 점수 (LGBM):', grid_cv_lgbm.best_score_)

# 최적 하이퍼파라미터로 모델 재학습 및 평가
best_lgbm = grid_cv_lgbm.best_estimator_

print('\n### 최적 하이퍼파라미터 LGBM 예측 성능 (SMOTE) ###')
get_model_train_eval(best_lgbm, ftr_train=X_train_over, ftr_test=X_test, tgt_train=y_train_over, tgt_test=y_test)

print('\n### 최적 하이퍼파라미터 LGBM F1 스코어 (SMOTE) ###')
best_lgbm_pred_proba = best_lgbm.predict_proba(X_test)[:, 1]
precision_recall_f1_plot(y_test, best_lgbm_pred_proba)
```

```python
print('\n### XGBoost 하이퍼파라미터 튜닝 ###')
# 하이퍼파라미터 튜닝을 위한 XGBoost 모델 생성
xgb_clf_tune = XGBClassifier(random_state=0, eval_metric='logloss')

# 튜닝할 하이퍼파라미터 그리드 설정
params = {
    'n_estimators': [200, 500],
    'max_depth': [5, 10],
    'learning_rate': [0.05, 0.1]
}

# GridSearchCV를 사용하여 최적의 하이퍼파라미터 탐색
grid_cv_xgb = GridSearchCV(xgb_clf_tune, param_grid=params, cv=3, scoring='f1', verbose=1, n_jobs=-1)
grid_cv_xgb.fit(X_train_over, y_train_over)

print('GridSearchCV 최적 하이퍼파라미터 (XGBoost):', grid_cv_xgb.best_params_)
print('GridSearchCV 최고 F1 점수 (XGBoost):', grid_cv_xgb.best_score_)

# 최적 하이퍼파라미터로 모델 재학습 및 평가
best_xgb = grid_cv_xgb.best_estimator_

print('\n### 최적 하이퍼파라미터 XGBoost 예측 성능 (SMOTE) ###')
get_model_train_eval(best_xgb, ftr_train=X_train_over, ftr_test=X_test, tgt_train=y_train_over, tgt_test=y_test)

print('\n### 최적 하이퍼파라미터 XGBoost F1 스코어 (SMOTE) ###')
best_xgb_pred_proba = best_xgb.predict_proba(X_test)[:, 1]
precision_recall_f1_plot(y_test, best_xgb_pred_proba)
```
