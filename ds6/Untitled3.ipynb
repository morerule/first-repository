```python
import random

class Account: # 은행 계좌를 관리하는 클래스입니다.

    # 클래스 변수: 생성된 계좌 객체의 개수 저장
    account_count = 0
    BANK_NAME = '모두은행'

    def __init__(self, depositor_name):
        self.depositor_name = depositor_name
        self.balance = 0
        self.account_number = self._generate_account_number()
        self.deposit_history = []
        self.withdrawal_history = []
        self.deposit_count = 0

        # 계좌 생성 시 계좌 개수 증가
        Account.account_count += 1
        print(f"새로운 계좌가 생성되었습니다. 예금주: {self.depositor_name}, 계좌번호: {self.account_number}")

    def _generate_account_number(self):
        """
        랜덤한 3-2-6 형식의 계좌번호를 생성합니다.
        """
        part1 = str(random.randint(100, 999))
        part2 = str(random.randint(10, 99))
        part3 = str(random.randint(100000, 999999))
        return f"{part1}-{part2}-{part3}"

    def deposit(self, amount):
        """
        계좌에 입금합니다. 최소 입금액은 1원입니다.
        """
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_history.append(amount)
            print(f"{amount}원이 입금되었습니다. 현재 잔액: {self.balance}원")

            if self.deposit_count % 5 == 0:
                interest = sum(self.deposit_history) * 0.01
                self.balance += interest
                print(f"5회 입금 달성! 이자 1% ({interest}원)가 지급되었습니다. 현재 잔액: {self.balance}원")
        else:
            print("입금은 최소 1원 이상만 가능합니다.")

    def withdraw(self, amount):
        """
        계좌에서 출금합니다. 잔액 이상은 출금할 수 없습니다.
        """
        if amount <= self.balance:
            self.balance -= amount
            self.withdrawal_history.append(amount)
            print(f"{amount}원이 출금되었습니다. 현재 잔액: {self.balance}원")
        else:
            print("잔액이 부족하여 출금할 수 없습니다.")

    @classmethod
    def get_account_num(cls):
        """
        생성된 총 계좌 개수를 출력합니다.
        """
        print(f"총 생성된 계좌 개수: {cls.account_count}개")

    def history(self):
        """
        입금 및 출금 내역을 출력합니다.
        """
        print("\n--- 거래 내역 ---")
        print("입금 내역:", self.deposit_history)
        print("출금 내역:", self.withdrawal_history)
        print("-----------------")


# --- 예제 코드 ---
if __name__ == "__main__":
    # 1. 계좌 생성
    print("--- 계좌 생성 ---")
    my_account = Account("홍길동")
    your_account = Account("김철수")

    # 2. 총 계좌 개수 확인
    print("\n--- 총 계좌 개수 확인 ---")
    Account.get_account_num()

    # 3. 입금 및 출금 테스트
    print("\n--- 거래 테스트 ---")
    my_account.deposit(10000)
    my_account.deposit(5000)
    my_account.deposit(2000)
    my_account.withdraw(1000)
    my_account.deposit(3000)
    my_account.deposit(1000) # 5회차 입금, 이자 지급

    # 4. 거래 내역 확인
    print("\n--- 거래 내역 확인 ---")
    my_account.history()
    your_account.history()
```

```python
import random

class Account:
    bank_name = "모두은행"
    account_count = 0

    def __init__(self, owner):
        self.owner, self.balance = owner, 0
        self.account_number = f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"
        self.log, self.deposit_count = [], 0
        Account.account_count += 1

    def _add_log(self, kind, amount):
        self.log.append(f"{len(self.log)+1:>2}회 | {kind:<4} | 금액: {amount:>11,}원 | 잔액: {self.balance:>11,}원")

    def deposit(self, amt):
        if amt < 1: return print("입금은 최소 1원 이상이어야 합니다.")
        self.balance += amt; self.deposit_count += 1; self._add_log("입금", amt)
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest; self._add_log("이자", interest)

    def withdraw(self, amt):
        if amt > self.balance: return print("잔액이 부족합니다.")
        self.balance -= amt; self._add_log("출금", amt)

    def __str__(self):
        return f"\n[{self.bank_name}] 계좌정보\n예금주: {self.owner}\n계좌번호: {self.account_number}\n잔액: {self.balance:,}원\n거래 내역:\n" + "\n".join(self.log or ["거래 내역 없음"])

accounts = {}

while True:
    name = input("\n이름 입력 (종료: exit): ").strip()
    if name.lower() == "exit": break
    acc = accounts.get(name) or Account(name); accounts[name] = acc
    print(f"\n{name}님 계좌 {'확인' if name in accounts else '생성'}됨\n{acc}")

    while True:
        cmd = input("\n작업 선택 (1/입금, 2/출금, 3/정보, 4/변경): ").strip()
        if cmd in ["1", "입금"]: acc.deposit(int(input("입금 금액: ")))
        elif cmd in ["2", "출금"]: acc.withdraw(int(input("출금 금액: ")))
        elif cmd in ["3", "정보"]: print(acc)
        elif cmd in ["4", "변경"]: break
        else: print("다시 입력해주세요.")
```
