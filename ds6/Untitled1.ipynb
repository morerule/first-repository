```python
# Cell 1 : ì •ë‹µ ìƒì„± + ì…ë ¥ ì²˜ë¦¬
import random

def generate_answer():  #  ê²Œì„ì„ ì‹œì‘í• ë•Œ ì •ë‹µ ìˆ«ì 5ê°œë¥¼ ë½‘ì•„ëŠ” í•¨ìˆ˜  #
    """
    ì„œë¡œ ë‹¤ë¥¸ 5ê°œì˜ ìˆ«ìë¥¼ ëœë¤ìœ¼ë¡œ ìƒì„±í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜
    """
    return random.sample(range(0, 10), 5)

def get_user_guess():
    """
    ì‚¬ìš©ìì—ê²Œ 5ìë¦¬ ìˆ«ìë¥¼ ì…ë ¥ë°›ì•„ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜
    """
    while True:
        guess = input("5ìë¦¬ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
        if guess.isdigit() and len(guess) == 5:
            return [int(d) for d in guess]
        else:
            print("ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•˜ì„¸ìš”.")
```

```python
# Cell 2 : íŒì • + ë©”ì¸ ë£¨í”„

def check_guess(answer, guess):
    """
    ì •ë‹µ(answer)ê³¼ ì…ë ¥(guess)ì„ ë¹„êµí•˜ì—¬
    strike, ball ê°œìˆ˜ë¥¼ ë°˜í™˜
    """
    strike = sum(a == g for a, g in zip(answer, guess))
    ball = sum((g in answer) and (answer[i] != g) for i, g in enumerate(guess))
    return strike, ball


# ë©”ì¸ ë£¨í”„ ì‹¤í–‰
answer = generate_answer()
print("ê²Œì„ ì‹œì‘! 5ìë¦¬ ìˆ«ìë¥¼ ë§ì¶°ë³´ì„¸ìš”.")

while True:
    guess = get_user_guess()
    strike, ball = check_guess(answer, guess)
    print(f"{strike} Strike, {ball} Ball")

    if strike == 5:
        print("ğŸ‰ ì •ë‹µì…ë‹ˆë‹¤! ê²Œì„ ì¢…ë£Œ!")
        break
```
